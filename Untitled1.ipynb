```python
import os
from glob import glob
import subprocess

# 원본 .wav 파일이 있는 폴더
input_dir = "/content/drive/MyDrive/merged_wav"
output_dir = "/content/drive/MyDrive/wav_16k"
os.makedirs(output_dir, exist_ok=True)

# 모든 .wav 파일 경로 수집
wav_files = glob(os.path.join(input_dir, "*.wav"))
print(f"총 {len(wav_files)}개 파일 변환 시작")
```

```python
for wav_path in wav_files:
    filename = os.path.basename(wav_path)
    output_path = os.path.join(output_dir, filename)

    subprocess.run([
        "ffmpeg", "-y",
        "-i", wav_path,
        "-ar", "16000",         # 샘플링 레이트 16kHz
        "-ac", "1",             # 모노 채널
        output_path
    ], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

print("✅ 샘플링 완료: 16kHz mono .wav")
```

```python
import torchaudio
import os
from glob import glob

# 확인할 폴더 경로 (샘플링된 .wav가 있는 곳)
wav_dir = "/content/drive/MyDrive/wav_16k"  # 너가 쓴 경로로 수정

# .wav 파일 목록 수집
wav_files = glob(os.path.join(wav_dir, "*.wav"))

# 점검
bad_files = []
for path in wav_files:
    try:
        waveform, sr = torchaudio.load(path)

        if sr != 16000:
            bad_files.append((os.path.basename(path), f"❌ 샘플링 레이트 = {sr}"))

        elif waveform.shape[0] != 1:
            bad_files.append((os.path.basename(path), "❌ 모노 채널 아님"))

    except Exception as e:
        bad_files.append((os.path.basename(path), f"❌ 로드 오류: {str(e)}"))

# 결과 출력
if bad_files:
    print(f"⚠️ 문제 있는 파일 {len(bad_files)}개 발견:")
    for name, reason in bad_files[:10]:
        print(f"- {name} → {reason}")
    if len(bad_files) > 10:
        print("...생략...")
else:
    print(f"✅ {len(wav_files)}개 파일 모두 정상 (16kHz, mono)")
```
